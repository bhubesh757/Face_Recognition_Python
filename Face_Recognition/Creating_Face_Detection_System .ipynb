# Face detection using opencv

```python
# => create a cascade classiifer , which contains the feautures of the face
# => search for the row and column valueas of the face numpy ndarray
# => displays the image with the rectangular face box
```

```python
# create a Cascadeclassifier object 
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# reading the image 
img = cv2.imread('Lionel_Messi_20180626.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
# reading the image as gray scale
gray_image = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

# Face detection

#### detectMultiscale module of the classifier.This function will return the co-ordinates(x and y posiiton plus the height and width) of the detected faces as Rect(x,y,w,h).

```python
faces = face_cascade.detectMultiScale(gray_image, scaleFactor = 1.2, minNeighbors = 5);
```

```python
# to find the no of faces in the image
print('Faces found ' , len(faces))
```

```python
# search for the coordintes ofthe gray
```

```python
print(type(faces))
```

```python
print(faces)
```

#### next step is to loop all the coordinates it returns and draw rectangle shape around the face using the opencv. 

```python
for x,y,w,h in faces:
    img = cv2.rectangle(img , (x,y) ,(x+w ,y+h) , (0,255,0),3)
    
```

```python
resized = cv2.resize(img , (int(img.shape[1]) , int(img.shape[0])))
```

```python
cv2.imshow("GRAY" , resized)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

# Capturing the video

```python
import cv2,time
video = cv2.VideoCapture(0)

a = 1
while True:
    a = a+1
    check,frame = video.read()
    print(frame)
    gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    cv2.imshow('capture' , gray)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
print(a)
video.release()
cv2.destroyAllWindows()
```
